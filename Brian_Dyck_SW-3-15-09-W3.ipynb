```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\Obyume\Documents\Anaconda Doc\Brian_Dyck_SW-3-15-09-W3.csv")
df
```

```python
# converting date columns to datetime and extracting by year
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Year'] = df['Date'].dt.year
```

```python
#calculating the yealry mean at each depth
grouped_mean = df.groupby(['Year', 'Depth_cm']).agg({'Moisture': 'mean',
    'NDVI': 'mean',
    'NDMI': 'mean',
    'SAVI': 'mean',
    'GCI': 'mean'
}).reset_index()
print(grouped_mean)
```

```python
#Calculating soil moisture statistics at each depth for each year
grouped_sm = df.groupby(['Year', 'Depth_cm'])['Moisture'].describe()
print(grouped_sm)
```

```python
# calculating indices statistics per year
grouped_indices = df.groupby('Year')[['NDVI', 'NDMI', 'SAVI', 'GCI']].describe()
print(grouped_indices)
```

```python
#Trend analysis for soil moisture
plt.figure(figsize=(14, 8))
for depth_cm in df['Depth_cm'].unique():
    print('depth_cm: ', depth_cm)
    subset = df[df['Depth_cm'] == depth_cm]
    print(subset)
    plt.plot(subset['Date'], subset['Moisture'], label=depth_cm)
    plt.scatter(subset['Date'], subset['Moisture'])

plt.xlabel('Date')
plt.ylabel('Moisture')
plt.title('Soil_Moisture Levels Over Time at Different Depth_cm')
plt.legend()
plt.show()
```

```python
#Trend analysis for Vegetative indices
plt.figure(figsize=(12, 8))

subset = df[df['Depth_cm'] == 0]  # Filter the DataFrame for the current depth_cm
print(subset)
for name in ['NDVI', 'NDMI', 'SAVI', 'GCI', 'Moisture']:
    plt.plot(subset['Date'], subset[name], label=f'{name}')
    plt.scatter(subset['Date'], subset[name])

plt.xlabel('Date')
plt.ylabel('Index Values')
plt.title('Vegetation Indices Over Time')
plt.legend()
plt.show()
```

```python
#Calculating and plotting Correlation coefficient of soil moisture at depth =0 with vegetative indices.
# Filter the data for Depth_cm = 0
df_Depth_cm_0 = df[df['Depth_cm'] == 0]
correlation_matrix = df_Depth_cm_0.corr()
# Extract correlations of Soil_Moisture with NDVI, NDMI, SAVI, and GCI
correlations = correlation_matrix.loc['Moisture', ['NDVI', 'NDMI', 'SAVI', 'GCI']]

print("Correlation coefficients for soil moisture at Depth_cm = 0 with vegetation indices:")
print(correlations)
```

```python
#Trend analysis for Vegetative indices
plt.figure(figsize=(12, 8))

subset = df[df['Depth_cm'] == 0]  # Filter the DataFrame for the current depth_cm
print(subset)

# Plot bars for Moisture and NDVI
plt.bar(subset['Date'], subset['Moisture'], width=7, label='Moisture', align='center', alpha=0.7, color='blue')
plt.bar(subset['Date'], subset['NDVI'], width=7, label='NDVI', align='edge', alpha=0.7, color='green')

# Plot lines for threshold values
plt.plot(subset['Date'], subset['Treshold_M'], label='Treshold_M', linestyle='-', color='blue')
plt.plot(subset['Date'], subset['NDVI_MAXTR'], label='NDVI_MAXTR', linestyle='-.', color='green')

# Optional: Scatter points for the thresholds (if needed)
plt.scatter(subset['Date'], subset['Treshold_M'], color='red')
plt.scatter(subset['Date'], subset['NDVI_MAXTR'], color='orange')

# Formatting the plot
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('NDVI_NW-18-9-18-W3 and Moisture trend Over Time with max threshold')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
```
